## Leetcode 238. Product of Array Except Self


Tính tích dồn 2 chiều

[Link bài tập](https://leetcode.com/problems/product-of-array-except-self/)

````python
from typing import List
````

**Example 1:**
```
Input: nums = [1,2,3,4]
Output: [24,12,8,6]
```
**Example 2:**
```
Input: nums = [-1,1,0,-3,3]
Output: [0,0,9,0,0]
```

**Constraints:**
```
2 <= nums.length <= 105
-30 <= nums[i] <= 30
The product of any prefix or suffix of nums is guaranteed to fit in a 32-bit integer.
```

````python
def productExceptSelf(nums: List[int]) -> List[int]:
    # Lập trình tại đây
    length = len(nums)
    result = [1] * length
 
    for i in range(1, length):
        result[i] = result[i - 1] * nums[i - 1]
        
    right = nums[-1]

    for i in range(length - 2, -1, -1):
        result[i] *= right
        right *= nums[i]

    return result
````
