## 442. Find All Duplicates in an Array

Tìm mảng những giá trị lặp lại

[Link bài tập](https://leetcode.com/problems/find-all-duplicates-in-an-array/)

````python
from typing import List
````

**Example 1:**
```
Input: nums = [4,3,2,7,8,2,3,1]
Output: [2,3]
```
**Example 2:**
```
Input: nums = [1,1,2]
Output: [1]
```

**Constraints:**
```
n == nums.length
1 <= n <= 10**5
1 <= nums[i] <= n
Each element in nums appears once or twice.
```

````python
def findDuplicates(nums: List[int]) -> List[int]:
    # Lập trình tại đây
    result = []
    appreared = set()
    for num in nums:
        if num in appreared:
            result.append(num)
        else:
            appreared.add(num)
    return result
````

## 268. Missing Number

Tìm giá trị bị thiếu

[Link bài tập](https://leetcode.com/problems/missing-number/)


**Example 1:**

```
Input: nums = [3,0,1]
Output: 2
Explanation: n = 3 since there are 3 numbers, so all numbers are in the range [0,3]. 2 is the missing number in the range since it does not appear in nums.
```

**Example 2:**
```
Input: nums = [0,1]
Output: 2
Explanation: n = 2 since there are 2 numbers, so all numbers are in the range [0,2]. 2 is the missing number in the range since it does not appear in nums.
```


**Example 3:**
```
Input: nums = [9,6,4,2,3,5,7,0,1]
Output: 8
Explanation: n = 9 since there are 9 numbers, so all numbers are in the range [0,9]. 8 is the missing number in the range since it does not appear in nums.
```

**Constraints:**
```
n == nums.length
1 <= n <= 10**4
0 <= nums[i] <= n
All the numbers of nums are unique.

```

````python
def missingNumber(nums: List[int]) -> int:
    # Lập trình tại đây
    n = len(nums)
    return n*(n+1)//2-sum(nums)
````

Tìm tất cả giá trị bị biến mất

[Link bài tập](https://leetcode.com/problems/find-all-numbers-disappeared-in-an-array/)

```
Input: nums = [4,3,2,7,8,2,3,1]
Output: [5,6]
```

**Example 2:**
```
Input: nums = [1,1]
Output: [2]
```




**Constraints:**
```
n == nums.length
1 <= n <= 10**5
1 <= nums[i] <= n
```

````python
def findDisappearedNumbers(nums: List[int]) -> List[int]:
    # Lập trình tại đây
    nums.sort()
    n = len(nums)
    nums = set(nums)
    print(nums)
    result = []
    for i in range(1, n):
        if i+1 != nums[i]:
            result.append(i+1)
    if n not in nums:
        result.append(n)
    return result
````
