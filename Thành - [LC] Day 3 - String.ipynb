## 1. Đảo ngược chuỗi

Tiến hành đảo ngược chuỗi bất kỳ

````python
from typing import List
````

**Example 1:**

```
Input: s = "abc"
Output: "cba"
```


**Example 2:**

```
Input: s = "abcd"
Output: "dcba"

```

**Constraints:**
```
1 <= s.length <= 1000
```

````python
def reverseString(s: str) -> str:
    # Lập trình tại đây
    i = 0
    j = len(s) - 1
    while i < j:
        s = list(s)
        s[i], s[j] = s[j], s[i]
        i += 1
        j -= 1
    return ''.join(s)
````

## 2. Leetcode 345 - Reverse Vowels of a String

Đảo ngược nguyên âm của một chuỗi

[Link bài tập](https://leetcode.com/problems/reverse-vowels-of-a-string/)


**Example 1:**

```
Input: s = "hello"
Output: "holle"
```

**Example 2:**
```
Input: s = "leetcode"
Output: "leotcede"
```


**Constraints:**
```
1 <= s.length <= 3 * 105
s consist of printable ASCII characters.

```

````python
def reverseVowels(s: str) -> str:
    # Lập trình tại đây
    i = 0
    j = len(s) - 1
    s = list(s)
    while i < j:
        if s[i] in 'aeiouAEIOU':
            while j > i:
                if s[j] in 'aeiouAEIOU':
                    s[i], s[j] = s[j], s[i]
                    j -= 1
                    break
                j -= 1
        i += 1
    return ''.join(s)
````

## 3. Đếm số lượng chuỗi con chỉ có duy nhất một loại chữ



**Example 1:**

```
Input: s = "aaaba"
Output: 8

Giải thích: Các chuỗi con với duy nhất 1 loại chữ đó là "aaa", "aa", "a", "b" và
- "aaa" xuất hiện 1 lần
- "aa" xuất hiện 2 lần
- "a" xuất hiện 4 lần
- "b" xuất hiện 1 lần

Tổng số lần xuất hiện là 1 + 2 + 4 + 1 = 8 lần
```

**Ràng buộc:**
```
1 <= s.length <= 1000
Chuỗi chỉ chứa chữ cái tiếng Anh viết thường.
```

````python
def countSubStringTimes(s: str) -> int:
    # Lập trình tại đây
    result = 0
    i = j = 0
    while i < len(s):
        if s[i] == s[j]:
            result += 1
            # print(s[i:j+1])
        else:
            i += 1
            j = i
            continue
        if j + 1 <= len(s) - 1:
            j += 1
        else:
            i += 1
            j = i
        
    return result
````

## 4. Leetcode 541 - Đảo ngược chuỗi 2

Đảo ngược k kí tự trong 2k ký tự bắt đầu của chuỗi

[Link bài tập](https://leetcode.com/problems/reverse-string-ii/)



**Example 1:**

```
Input: s = "abcdefg", k = 2
Output: "bacdfeg"
```

**Example 2:**

```
Input: s = "abcd", k = 2
Output: "bacd"
```

**Constraints:**

```
1 <= s.length <= 104
s consists of only lowercase English letters.
1 <= k <= 104
```

````python
def reverse(s: str) -> str:
    i = 0
    j = len(s) - 1
    while i < j:
        s = list(s)
        s[i], s[j] = s[j], s[i]
        i += 1
        j -= 1
    return ''.join(s)
    
def reverseStr(s: str, k: int) -> str:
    # Lập trình tại đây
    result = ''
    e = k
    while e <= len(s):
        result += reverse(s[e-k:e])
        result += s[e:e+k]
        e += 2*k
    if e - k < len(s):
        result += reverse(s[e-k:])
    return result
````

## 5. Leetcode - 1347. Minimum Number of Steps to Make Two Strings Anagram

Số bước nhỏ nhất để tạo ra 2 chuỗi Anagram.

[Link bài tập](https://leetcode.com/problems/minimum-number-of-steps-to-make-two-strings-anagram/)

**Example 1:**

```
Input: s = "bab", t = "aba"
Output: 1
Explanation: Replace the first 'a' in t with b, t = "bba" which is anagram of s.
```

**Example 2:**

```
Input: s = "leetcode", t = "practice"
Output: 5
Explanation: Replace 'p', 'r', 'a', 'i' and 'c' from t with proper characters to make t anagram of s.
```
**Example 3:**

```
Input: s = "anagram", t = "mangaar"
Output: 0
Explanation: "anagram" and "mangaar" are anagrams. 
```

**Constraints:**

```
1 <= s.length <= 5 * 104
s.length == t.length
s and t consist of lowercase English letters only.
```

````python
def minSteps(s: str, t: str) -> int:
    result = 0
    # Lập trình tại đây
    for char in set(s):
        if s.count(char) > t.count(char):
            result += (s.count(char) - t.count(char))
    return result
````

## 6. Leetcode 125 - Valid Palindrome

Làm sạch chuỗi và kiểm tra chuỗi Palindrome.

[Link bài tập](https://leetcode.com/problems/valid-palindrome/)

**Example 1:**

```
Input: s = "A man, a plan, a canal: Panama"
Output: true
Explanation: "amanaplanacanalpanama" is a palindrome.
```



**Example 2:**

```
Input: s = "race a car"
Output: false
Explanation: "raceacar" is not a palindrome.
```

**Example 3:**
```
Input: s = " "
Output: true
Explanation: s is an empty string "" after removing non-alphanumeric characters.
Since an empty string reads the same forward and backward, it is a palindrome.
```

**Constraints:**

```
1 <= s.length <= 2 * 105
s consists only of printable ASCII characters.
```

````python
def isPalindrome(s: str) -> bool:
    """
    :type s: str
    :rtype: bool
    """
    s_new = ""
    for char in s:
        if char.isalpha() or char.isdigit():
            s_new = s_new + char.lower()
    if len(s_new) <= 1:
        return True
    s = s_new
    i = 0
    j = len(s) - 1
    while s[i] == s[j] and i < j:
        i += 1
        j -= 1
        if i >= j:
            return True
    return False  
````
